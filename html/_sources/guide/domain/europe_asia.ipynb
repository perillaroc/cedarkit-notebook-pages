# 欧亚大陆

欧亚大陆底图布局 (`cedarkit.maps.domains.EuropeAsiaMapTemplate`) 绘制欧亚大陆区域图形，使用兰勃特等角圆锥投影 (`cartopy.crs.LambertConformal`)，中心经度东经 95 度。

## 准备

导入需要的包

```python
import numpy as np
import pandas as pd

from cedarkit.maps.chart import Panel
from cedarkit.maps.domains import EuropeAsiaMapTemplate
from cedarkit.maps.map import set_default_map_loader_package
from cedarkit.maps.colormap import get_ncl_colormap
from cedarkit.maps.style import ContourStyle
```

设置使用 CEMC 地图包

```python
set_default_map_loader_package("cedarkit.maps.map.cemc")
```

定义绘图样式

```python
prmsl_color_map = get_ncl_colormap(
    "rainbow+white+gray",
    index=np.array([-1, 80, 100, 120, 140, 160, 180, 200, 205])
)
prmsl_contour_lev = np.arange(1000, 1036, 5)
prmsl_style = ContourStyle(
    colors=prmsl_color_map,
    levels=prmsl_contour_lev,
    fill=True,
)
```

## 绘制

绘制底图布局

```python
domain = EuropeAsiaMapTemplate()
panel = Panel(domain=domain)

domain.set_title(
    panel=panel,
    graph_name="500 hPa HGT (10gpm) line + MSLP (hPa) shadow",
    system_name="CMA-GFS",
    start_time=pd.to_datetime("2024-03-21 00:00"),
    forecast_time=pd.to_timedelta("24h"),
)
domain.add_colorbar(panel=panel, style=prmsl_style)

panel.show()
```
